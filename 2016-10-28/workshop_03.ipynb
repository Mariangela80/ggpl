## Implementations  
Both models have been created in three different ways.  
This has led to **_three slightly different solutions_** for each model, and to **_a total of six solutions_**.  

## Models specifications
### Model 1
1. U shaped stairs always starts with a **_first left fligh of stairs_**, so producing a **_clockwise ascent direction_**. 
2. The **_last step_** of each flight of stairs is **_on the same plane of the landing_** reached by the flight ; in particular, this last step is juxtaposed with the landing itself.  

### Model 2
1. U shaped stairs always starts with a **_first right fligh of stairs_**, so producing an **_anti-clockwise ascent direction_**.  
2. The **_last step_** of each flight of stairs is **_part of the reached landing_**. 

## Geometric methods  
The three solutions of each model (Model 1 and Model 2) have the first part of their geometric method in common.  
![Blondel formula](img/BlondelFormula.jpg "Blondel formula")
1. It uses
   * the **_Blondel's formula_**:   
     $2A + P = 63 cm$  
     and
   * the **_relationship of similarity_** between the stair's rise (dz) and run (dy), and the step's riser height (A) and thread depth (P):    
   $A/P = dz/dy$  

   to calculate 
   * the number of steps:    
     $steps = {round}(dz/A)$  
     and  
   * the step's dimensions:  
     $riser = dz/steps$  
     $thread = dy/steps$  
   
2. It fulfills **_two stair design rules_**:  
   * the **_stair's slope_** must be between 28 and 36 degrees:  
     $28 <= pitch <= 36$
   * the **_landing's depth_** must amount to the flight's width al least:  
   ![Dimensions Relationship](img/dimensionsRelationship.png "Dimension relationship")   
   

### 1) First Solution's geometric method 
This is **the best and shorter solution**.  

* A flight of stairs and its relative landing (the landing reached by it) are assembled into an unique Hpc (by means of the Plasm primitive STRUCT).  
* Two reference flights of stairs (and their relative landings) are created: the left and the right one.  
In particular, the right flight is created reflecting and properly translating the left one (by means of the affine transformations' Plasm primitives S and T respectively).  
* Different instances of these reference left and right flights of stairs are properly translated along the y and z directions and assembled together, so composing the final Hpc corresponding to the U shaped stair.  
* Note that the first left flight of stairs is different from the other left flights, because its first step mustn't be connected to any other previous landing.  


### 2) Second Solution's geometric method
It has been implemented just for exercise, with the aim of trying a **_pairwise replication_** of left and right flights (and their relative reached landings).
This solution revealed itself a little bit trickier than the simpler and shorter first one.  

Such as in the first solution:  
* A flight of stairs and its relative landing (the landing reached by it) are assembled into an unique Hpc (by means of the Plasm primitive STRUCT).  
* Two reference flights of stairs (and their relative landings) are created: the left and the right one.  
In particular, the right flight is created reflecting and properly translating the left one (by means of the affine transformations' Plasm primitives S and T respectively).  

In addition:  
* The left and right reference flights of stairs (and their relative reached landings) are assembled, together with the proper translation tensors along the y and z directions, into a unique Hpc representing the reference "double flight".  
* Different instances of the reference "double flight" are properly translated along the z direction and assembled together, so composing the final Hpc corresponding to the U shaped stair.  
* Note that the first "double flight" of stairs is different from the other ones, because the first step of the first left flight of stairs mustn't be connected to any other previous landing.   


### 3) Third Solution's geometric method
This third solution has been implemented in order **_to maintain the entity "landings" separated from the flights of stairs themselves_**.  

It has been implemented in two ways: starting from both first and second solution.  

* This time each flight of stairs is considered independently from its relative landing.   

**First way (starting from first solution)**
* Two reference flights  of stairs (this time without their relative landings) are created: the left and the right one.  
In particular, the right flight is created reflecting and properly translating the left one (by means of the affine transformations' Plasm primitives S and T respectively). 
* Different instances of these reference left and flight flights of stairs are properly translated along the y and z directions and assembled together, so composing the Hpc corresponding to all flights of stairs without landings.    
* Again, note that the first left flight of stairs is different from the other left flights, because its first step mustn't be connected to any other previous landing.  
* All landings are assembled together, so composing the Hpc corresponding to all landings.  
* The Hpc of all flights of stairs and the Hpc of all landings are assembled together, so composing the final Hpc corresponding to the U shaped stair.  

**Second way (starting from second solution)**   

Such as in the first solution:  
* Two reference flights  of stairs (this time without their relative landings) are created: the left and the right one.  
In particular, the right flight is created reflecting and properly translating the left one (by means of the affine transformations' Plasm primitives S and T respectively).  
    
In addition:  
* The left and right reference flights of stairs (this time without their relative reached landings) are assembled, together with the proper translation tensors along the y and z directions, into a unique Hpc representing the reference "double flight".  
* Different instances of the reference "double flight" are properly translated along the z direction and assembled together, so composing the Hpc corresponding to all flights of stairs without landings.    
* Again, note that the first "double flight" of stairs is different from the other ones, because the first step of the first left flight of stairs mustn't be connected to any other previous landing. 
* All landings are assembled together, so composing the Hpc corresponding to all landings.  
* The Hpc of all flights of stairs and the Hpc of all landings are assembled together, so composing the final Hpc corresponding to the U shaped stair. 


### Model 1 Specific geometry  
Let's remember Model 1 specifications:  
* U shaped stairs always starts with a **_first left fligh of stairs_**, so producing a **_clockwise ascent direction_**. 
* The **_last step_** of each flight of stairs is **_on the same plane of the landing_** reached by the flight; in particular, this last step is juxtaposed with the landing itself.  

By construction, the height of the stair (`dz` if the U shaped stair consists of a single flight (single floor), `dz/n` if it is made up of `n` flights (`n` floors)) is entirely spanned by a flight of stairs. In other words, the height of the stair includes the reached landing's thickness, always fixed at the step's riser height.   
It means that the last step of the considered flight of stairs is automatically at the same height, that is on the same plane, of the reached landing.  
  
**_Fixing the landing's depth at the flight's width_**, and calculating the stair spanning the y distance **`dy - 2*landingDepth`**, the specifications are automatically fulfilled.


### Model 2 Specific geometry
Let's remember Model 2 specifications:
* U shaped stairs always starts with a **_first right fligh of stairs_**, so producing an **_anti-clockwise ascent direction_**.  
* The **_last step_** of each flight of stairs is **_part of the reached landing_**. 

The stair is calculated in the same way, that is it is considered spanning the entire stair's height (i.e. its height includes the reached landing's thickness) and the entire y distance `dy - 2*landingDepth`, but:
* considering that the landing thickness amounts to the step's riser height, and
* with the aim of letting the last step to be part or the reached landing,  

**_just a one step less stair is displayed_**.  
**_The missing step's thread depth is included in the starting and reached landing_**, whose **_width_** this time is so **_fixed at the sum of the flight's width and the missing step's half thread depth_**.  

In order to obtain a stair starting with a right flight of stairs, the **_Model 1 U shaped stair is properly reflected with respect to the x axis and translated along the x axis itself_**.

## Code & Output (Model 1)

### Generation of a step (Model 1)

#### step(width,thread,riser,typology)  

This function builds and returns the Hpc of the step described by the input parameters: width, thread depth, riser height and typology.  
In particular, it can create three different kinds of step:  
1. `typology=0`: "**initial**" step, i.e. the first step of all flights of stairs starting from a landing (not for the first flight)  
2. `typology=1`: "**normal**" step, i.e. the first step of the first flight of stairs and all intermediate steps  
3. `typology=2`: "**final**" step, i.e. the last step of a flight of stairs, on the same plane of the reached landing.

````python
from larlib import *
from math import *

def step(width,thread,riser,typology):
    """
    Builds and returns the Hpc of a step
    
    width: step width
    thread: step depth
    riser: step height
    typology: typology of the step; 0 for "initial" step (the first step of all flights of stairs 
                                      starting from a landing (not for the first flight))
                                    1 for "normal" step (the first step of the first flight of
                                      stairs and all intermediate steps)
                                    2 for "final" step (the last step of a flight of stairs, on
                                      the same plane of the reached landing)
    """
    if (typology==0):    # initial step
        section = MKPOL([[[0.0,riser],[2*thread,riser],[0,-riser]],[[1,2,3]],None])
        step = PROD([section,Q(width)])
        step = MAP([S3,S1,S2])(step)
    elif (typology==1):  # normal step
        section = MKPOL([[[0.0,0.0],[thread,0.0],[2*thread,riser],[0.0,riser]],[[1,2,3,4]],None])
        step = PROD([section,Q(width)])
        step = MAP([S3,S1,S2])(step)
    else:                # final step
        step = CUBOID([width,thread,riser])
        
    return step
````

### Generation of a flight  (Model 1)

#### flight (width,thread,riser,stepsNumber,firstStepType)  

This function builds and returns the Hpc ot the flight of stairs described by the input parameters: width, thread depth, riser height, number of steps, typology of the first step (0 for "initial" step, 1 for "normal" step, that is 0 for the first step of the first flight, and 1 for the first step of all flights starting from a landing).  
The last step is juxtaposed with the reached landing.  

````python
def flight (width,thread,riser,stepsNumber,firstStepType):
    """
    Builds and returns the Hpc of a flight of stairs
    
    width: step (flight) width
    thread: step depth
    riser: step height
    stepsNumber: number of steps in the flight of stairs
    firstStepType: typology of the first step in the flight of stairs;
                   1 for "normal" initial step, i.e. for the first step of the first flight
                   0 for "initial" step, i.e. for the first step of all flights of stairs starting 
                     from a landing (not for the first flight)
    """
    
    firstStep = step(width,thread,riser,firstStepType)
    normalStep = step(width,thread,riser,1)
    lastStep = step(width,thread,riser,2)
    
    flight = [firstStep,T([2,3])([thread,riser])]
    flight += NN(stepsNumber-2)([normalStep,T([2,3])([thread,riser])])
    flight += [lastStep]
    
    return STRUCT(flight)
````

### Generation  of an U shaped straight stair with half landings  (Model 1) (first, second and third solutions)

#### ggpl_U_shaped_stair (dx,dy,dz,floors=2,stairwell=0)

This function builds and returns the Hpc of the U shaped stair described by the input parameters.  
In particular, it accepts as input the x, y and z dimensions of the stair's containment box, the number of floors (default 2) and the x dimension of the stairwell (default 0).  

In this way:
* **the floors' number isn't fixed, but it can be decided from the outside**
* **it contemplates the space for a potential stairwell**

Note that, in particular, **the function returns the STRUCT of the U shaped stair and the 1 skeleton of the containment box described by the input, in order to show that the built model fits in the input dimensions perfectly, with no need of additional rescaling**.

### Generation  of an U shaped straight stair with half landings  (Model 1, first solution)

#### ggpl_U_shaped_stair (dx,dy,dz,floors=2,stairwell=0)

````python
def ggpl_U_shaped_stair(dx,dy,dz,floors=2,stairwell=0):
    """ 
    Builds and returns the Hpc of an U shaped stair
    
    dx,dy,dz: x, y and z dimensions of the stair's containment box
    floors: number of floors (default 2)
    stairwell: x dimension of the stairwell (default 0)
    """
    
    dx = float(dx)
    dy = float(dy)
    dz = float(dz)
    
    # checks if the input floors' number is integer
    if (not float(floors).is_integer()):
        raise Exception ("Wrong input type","The number of floors must be an integer")
    
    flightWidth = (dx-stairwell)/2
    landingDepth = flightWidth # the landing's depth must amount to the flight's width al least
    flightLength = dy-2*landingDepth
    flightHeight = dz/floors
    
    # checks if the stair's slope is between 28 and 36 degrees
    riserThreadRatio = flightHeight/flightLength
    pitch = degrees(ATAN(riserThreadRatio))

    # print pitch
    if (not (28 <= pitch <= 36)):
        raise Exception ("Wrong input value",
                         "Each flight must have a pitch between 28 and 36 degrees. "+
                         "The pith of your flight is " + str(pitch) +" degrees. "+
                         "Correct the dimensions of your stair.")
    
    # calculates the number of steps in each flight, the riser height and the thread depth
    thread = 63/(2*riserThreadRatio+1)
    riser = riserThreadRatio*thread
    
    steps = round(flightHeight*100/riser)
    riser = flightHeight/steps
    thread = flightLength/steps
     
    #-----------------------------------------------------------------------------------------------
    #FIRST SOLUTION
    
    landing = CUBOID([dx,landingDepth,riser])
    firstLeftFlightWithLanding = STRUCT([flight(flightWidth,thread,riser,steps,1),
                                       T([2,3])([flightLength,flightHeight-riser]),landing])
    leftFlightWithLanding = STRUCT([flight(flightWidth,thread,riser,steps,0),
                                  T([2,3])([flightLength,flightHeight-riser]),landing])
    rightFlightWithLanding = T([1,2])([dx,dy-landingDepth])(S([1,2])([-1,-1])(leftFlightWithLanding))
    
    uShapedStair = [T(2)(landingDepth),firstLeftFlightWithLanding,T(3)(flightHeight)]
    
    for floor in range(2,floors+1): # from 2 to floors
        if (floor%2!=0): # odd landing, reached by a left flight
            uShapedStair += [T(2)(landingDepth),leftFlightWithLanding,T(3)(flightHeight)]
        else:            # even landing, reached by a right flight
            uShapedStair += [T(2)(-landingDepth),rightFlightWithLanding,T(3)(flightHeight)]
            
    uShapedStair = STRUCT(uShapedStair)
    #-----------------------------------------------------------------------------------------------

    #print SIZE ([1,2,3]) (uShapedStair)
    #print SIZE ([1,2,3]) (BOX([1,2,3])(uShapedStair))
    
    containmentBox = SKEL_1(CUBOID( [dx,dy,dz] ))
    
    return STRUCT([containmentBox,uShapedStair])
````

````python
def ggpl_U_shaped_stair(dx,dy,dz,floors=2,stairwell=0):
    """ 
    Builds and returns the Hpc of an U shaped stair
    
    dx,dy,dz: x, y and z dimensions of the stair's containment box
    floors: number of floors (default 2)
    stairwell: x dimension of the stairwell (default 0)
    """
    
    dx = float(dx)
    dy = float(dy)
    dz = float(dz)
    
    # checks if the input floors' number is integer
    if (not float(floors).is_integer()):
        raise Exception ("Wrong input type","The number of floors must be an integer")
    
    flightWidth = (dx-stairwell)/2
    landingDepth = flightWidth # the landing's depth must amount to the flight's width al least
    flightLength = dy-2*landingDepth
    flightHeight = dz/floors
    
    # checks if the stair's slope is between 28 and 36 degrees
    riserThreadRatio = flightHeight/flightLength
    pitch = degrees(ATAN(riserThreadRatio))

    # print pitch
    if (not (28 <= pitch <= 36)):
        raise Exception ("Wrong input value",
                         "Each flight must have a pitch between 28 and 36 degrees. "+
                         "The pith of your flight is " + str(pitch) +" degrees. "+
                         "Correct the dimensions of your stair.")
    
    # calculates number of steps in each flight, riser height and thread depth
    
    thread = 63/(2*riserThreadRatio+1)
    riser = riserThreadRatio*thread
    
    steps = round(flightHeight*100/riser)
    riser = flightHeight/steps
    thread = flightLength/steps
       
    #-----------------------------------------------------------------------------------------------
    #SECOND SOLUTION

    landing = CUBOID([dx,landingDepth,riser])
    firstLeftFlightWithLanding = STRUCT([flight(flightWidth,thread,riser,steps,1),
                                       T([2,3])([flightLength,flightHeight-riser]),landing])
    leftFlightWithLanding = STRUCT([flight(flightWidth,thread,riser,steps,0),
                                  T([2,3])([flightLength,flightHeight-riser]),landing])
    rightFlightWithLanding = T([1,2])([dx,dy-landingDepth])(S([1,2])([-1,-1])(leftFlightWithLanding))
    
    uShapedStair = []
    doubleFloors = floors//2 # number of pairs of floors
    if (doubleFloors!=0):    # two floors at least, that is two flights (left and right) and two
                             # landings at least
        firstDoubleFlight = STRUCT([T(2)(landingDepth),firstLeftFlightWithLanding,
                                    T([2,3])([-landingDepth,flightHeight]),rightFlightWithLanding])
        uShapedStair = [firstDoubleFlight,T(3)(flightHeight*2)]
        doubleFloors = doubleFloors-1

        if(doubleFloors!=0): # two other floors at least =>
                             # `flightWithLanding`s' pairwise replication
            doubleFlight = STRUCT([T(2)(landingDepth),leftFlightWithLanding,
                                   T([2,3])([-landingDepth,flightHeight]),
                                   rightFlightWithLanding,T(3)(flightHeight)])   
            doubleFlights = [doubleFlight,T(3)(2*flightHeight)]*doubleFloors
            uShapedStair += doubleFlights
            
    if (floors%2!=0):        # if the floors' number is odd, we have to add the last left flight
                             # (and the relative reached landing)
        if(floors!=1):       # if there isn't only a floor, add the last 'leftFlightWithLanding`
            uShapedStair += [T(2)(landingDepth),leftFlightWithLanding] 
        else:                # if ther is only a floor, add the first and sole left flight (and 
                             # its relative reached landing)
            uShapedStair += [T(2)(landingDepth),firstLeftFlightWithLanding]
            
    uShapedStair = STRUCT(uShapedStair)
    #-----------------------------------------------------------------------------------------------
    
    #print SIZE ([1,2,3]) (uShapedStair)
    #print SIZE ([1,2,3]) (BOX([1,2,3])(uShapedStair))
    
    containmentBox = SKEL_1(CUBOID( [dx,dy,dz] ))
    
    return STRUCT([containmentBox,uShapedStair]) 
````

### Generation  of an U shaped straight stair with half landings  (Model 1, third solution)    

#### landings (length,width,height,flightHeight,flightLength,floors)  
This function builds and returns the Hpc corresponding to all landings in the U shaped stair described by the input parameters.  
In particular, it accepts as input the landings' length, width and thickness (i.e the x,y and z dimensions of each landing, respectively), the height and the length of each flight of stairs connecting landings, and the number of landings themselves reached by the U shaped stair (i.e. the number of floors).  

* **This functions has been added in order to let the solution be more modular.** 
______________________________________________________________________________________________________________

#### ggpl_U_shaped_stair(dx,dy,dz,floors=2,stairwell=0) 
This function, such as detailed in the first introductive section, is **implemented in two ways**:  
1. first way: **starting from first solution**
2. second way: **starting from second solution**

REFER TO SECTION ON "GEOMETRIC METHODS" FOR IMPLEMENTATION DETAILS AND DIFFERENCES BETWEEN THE FIRST AND SECOND IMPLEMENTATION WAYS; that section contains an exhaustive description.  

````python
def landings (length,width,height,flightHeight,flightLength,floors):
    """
    Builds and returns all landings
    
    length: landings' length (x dimension)
    width: landings' width (y dimension)
    height: landings' thickness (z dimension)
    flightHeight: the height (z dimension) between each landing (spanned by each flight of stairs)
    flightLength: the distance (y dimension) between each landing (i.e. the length of each flight
                  of stairs, leading from one landing to the following one)  
    floors: number of floors
    """
    
    landing = CUBOID([length,width,height])
    landings = [T(3)(-height)]
    
    for floor in range(1,floors+1): # from 1 to floors
        if (floor%2!=0): # odd landing, reached by a left flight
            landings += [T([2,3])([width+flightLength,flightHeight]),landing]
        else:            # even landing, reached by a right flight
            landings += [T([2,3])([-width-flightLength,flightHeight]),landing]
       
    return STRUCT(landings)
````

## Code & Output (Model 2)

### Generation of a step (Model 2)

#### step(width,thread,riser,typology)  

### Generation of a flight (Model 2)

#### flight (width,thread,riser,stepsNumber,firstStepType)

These functions are slightly different from the ones in Model 1.  
*  The function **step** just considers that there is **no need of type 2 step** because the last step of each flight of stairs isn't juxtaposed with the landing (such as in Model 1) anymore, but now it is part of the reached landing itself. 
* The function **flight** just considers that the last step in each flight of stairs isn't diffferent from the other ones (such as in Model 1), but it is precisely equal to each intermediate step.  
* The function **landings** is the same in Model 1, so it isn't replicated here either.  
   
REFER TO MODEL 1 FOR FURTHER DETAILS ON THESE FUNCTIONS.

````python
from larlib import *
from math import *

def step(width,thread,riser,typology):
    """
    Builds and returns the Hpc of a step
    
    width: step width
    thread: step depth
    riser: step height
    typology: typology of the step; 0 for initial step (the first step of flights of stairs 
                                      starting from a landing (not for the first flight))
                                    1 for normal step (the first step of the first flight of
                                      stairs and all intermediate steps)
    """
    if (typology==0):    # initial step
        section = MKPOL([[[0.0,riser],[2*thread,riser],[0,-riser]],[[1,2,3]],None])
        step = PROD([section,Q(width)])
        step = MAP([S3,S1,S2])(step)
    else:   # typology==1  normal step
        section = MKPOL([[[0.0,0.0],[thread,0.0],[2*thread,riser],[0.0,riser]],[[1,2,3,4]],None])
        step = PROD([section,Q(width)])
        step = MAP([S3,S1,S2])(step)
        
    return step


def flight (width,thread,riser,stepsNumber,firstStepType):
    """
    Builds and returns the Hpc of a flight of stairs
    
    width: step (flight) width
    thread: step depth
    riser: step height
    stepsNumber: number of steps in the flight of stairs
    firstStepType: typology of the first step in the flight of stairs;
                   1 for "normal" initial step, i.e. for the first step of the first flight
                   0 for "initial" step, i.e. for the first step of all flights of stairs starting 
                     from a landing (not for the first flight)
    """
    
    firstStep = step(width,thread,riser,firstStepType)
    normalStep = step(width,thread,riser,1)
    
    flight = [firstStep,T([2,3])([thread,riser])]
    flight += NN(stepsNumber-1)([normalStep,T([2,3])([thread,riser])])
    
    return STRUCT(flight)
````

````python
def ggpl_U_shaped_stair(dx,dy,dz,floors=2,stairwell=0):
    """ 
    Builds and returns the Hpc of an U shaped stair
    
    dx,dy,dz: x, y and z dimensions of the stair's containment box
    floors: number of floors (default 2)
    stairwell: x dimension of the stairwell (default 0)
    """
    
    dx = float(dx)
    dy = float(dy)
    dz = float(dz)
    
    # checks if the input floors' number is integer
    if (not float(floors).is_integer()):
        raise Exception ("Wrong input type","The number of floors must be an integer")
    
    flightWidth = (dx-stairwell)/2
    landingDepth = flightWidth # the landing's depth must amount to the flight's width al least
    flightLength = dy-2*landingDepth
    flightHeight = dz/floors
    
    # checks if the stair's slope is between 28 and 36 degrees
    riserThreadRatio = flightHeight/flightLength
    pitch = degrees(ATAN(riserThreadRatio))

    # print pitch
    if (not (28 <= pitch <= 36)):
        raise Exception ("Wrong input value",
                         "Each flight must have a pitch between 28 and 36 degrees. "+
                         "The pith of your flight is " + str(pitch) +" degrees. "+
                         "Correct the dimensions of your stair.")
    
    # calculates number of steps in each flight, riser height and thread depth
    
    thread = 63/(2*riserThreadRatio+1)
    riser = riserThreadRatio*thread
    
    steps = round(flightHeight*100/riser)
    riser = flightHeight/steps
    thread = flightLength/steps

    landingDepth += thread/2
       
    #-----------------------------------------------------------------------------------------------
    #SECOND SOLUTION
    
    landing = CUBOID([dx,landingDepth,riser])
    firstLeftFlightWithLanding = STRUCT([flight(flightWidth,thread,riser,steps-1,1),
                                       T([2,3])([flightLength-thread,flightHeight-riser]),landing])
    leftFlightWithLanding = STRUCT([flight(flightWidth,thread,riser,steps-1,0),
                                  T([2,3])([flightLength-thread,flightHeight-riser]),landing])
    rightFlightWithLanding = T([1,2])([dx,dy-landingDepth])(S([1,2])([-1,-1])(leftFlightWithLanding))
    
    uShapedStair = []
    doubleFloors = floors//2 # number of pairs of floors
    if (doubleFloors!=0):    # two floors at least, that is two flights (left and right) and two
                             # landings at least
        firstDoubleFlight = STRUCT([T(2)(landingDepth),firstLeftFlightWithLanding,
                                    T([2,3])([-landingDepth,flightHeight]),rightFlightWithLanding])
        uShapedStair = [firstDoubleFlight,T(3)(flightHeight*2)]
        doubleFloors = doubleFloors-1

        if(doubleFloors!=0): # two other floors at least =>
                             # `flightWithLanding`s' pairwise replication
            doubleFlight = STRUCT([T(2)(landingDepth),leftFlightWithLanding,
                                   T([2,3])([-landingDepth,flightHeight]),
                                   rightFlightWithLanding,T(3)(flightHeight)])   
            doubleFlights = [doubleFlight,T(3)(2*flightHeight)]*doubleFloors
            uShapedStair += doubleFlights
            
    if (floors%2!=0):        # if the floors' number is odd, we have to add the last left flight
                             # (and the relative reached landing)
        if(floors!=1):       # if there isn't only a floor, add the last 'leftFlightWithLanding`
            uShapedStair += [T(2)(landingDepth),leftFlightWithLanding] 
        else:                # if ther is only a floor, add the first and sole left flight (and 
                             # its relative reached landing)
            uShapedStair += [T(2)(landingDepth),firstLeftFlightWithLanding]
            
    uShapedStair = STRUCT(uShapedStair)
    uShapedStair = STRUCT([T(1)(dx),S(1)(-1),uShapedStair])
    #-----------------------------------------------------------------------------------------------
    
    #print SIZE ([1,2,3]) (uShapedStair)
    #print SIZE ([1,2,3]) (BOX([1,2,3])(uShapedStair))
    
    containmentBox = SKEL_1(CUBOID( [dx,dy,dz] ))
    
    return STRUCT([containmentBox,uShapedStair]) 
````
